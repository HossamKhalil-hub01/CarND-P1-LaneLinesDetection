## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.



#Convert to HLS color space to perform color masking on yellow and white colors
hls = cv2.cvtColor(image,cv2.COLOR_RGB2HLS)

#White color mask
white_min = np.array([0, 200,0])
white_max = np.array([180, 255,80])
white_mask = cv2.inRange(hls, white_min, white_max)

#Yellow color mask

yellow_min = np.array([20, 0, 90])
yellow_max = np.array([30, 255, 255])
yellow_mask = cv2.inRange(hls, yellow_min, yellow_max)


# Calculate combined mask, and masked image
mask = cv2.bitwise_or(yellow_mask, white_mask)

plt.imshow(mask,cmap='gray')
plt.title('White and Yellow Mask')
plt.show()



#Smooth using gaussian blur
kernal_size = 5
blur = cv2.GaussianBlur (mask,(kernal_size,kernal_size),0)

#Edge Detection
lower_thresh = 50
higher_thresh = 150

edge = cv2.Canny(blur,lower_thresh,higher_thresh)

plt.imshow(edge,cmap ='gray')
plt.title('Edge Detection')
plt.show()




#Region selection
region_mask = image*0

width1 = 0.45
width2 = 0.05
height = 0.6


#Forming the shape of the area as a function of the size (To be suitable for most of the cases)
pt1 = (int(image.shape[1]/2 - width1*image.shape[1]),int(image.shape[0]))
pt4 = (int(image.shape[1]/2 + width1*image.shape[1]),int(image.shape[0]))

pt2 = (int(image.shape[1]/2 - width2*image.shape[1]) ,int(height*image.shape[0]))
pt3 = (int(image.shape[1]/2 + width2*image.shape[1]) ,int(height*image.shape[0]))

pts = np.array([[pt1,pt2, pt3, pt4]])


#Select region of interest
cv2.fillPoly(region_mask,pts, 255)

#Visualize the selectred region
masked_im = cv2.addWeighted(region_mask,0.8,image,1,0)

plt.imshow(masked_im)
plt.title('Selected Region Mask')
plt.show()


#Filtering using region of interest
masked_edge = cv2.bitwise_and(edge,region_mask[:,:,0])

plt.imshow(masked_edge,cmap='gray')
plt.title('Filtered Edges')
plt.show()

# Hough lines detection
threshold = 10
rho = 1
theta = np.pi /180
min_len = 2
max_gap = 200


lines = cv2.HoughLinesP(masked_edge,rho = rho, theta = theta,threshold = threshold, minLineLength= min_len, maxLineGap= max_gap)
detected_lines = image*0



#Filtering the lines

#Construct a refernce line
ref_line = image*0
ref_pts = [int(image.shape[1]/2),int(image.shape[0]),int(image.shape[1]/2),int(height*image.shape[0])]

cv2.line(ref_line,(ref_pts[0],ref_pts[1]),(ref_pts[2],ref_pts[3]),255,10)


ref_line = cv2.addWeighted (ref_line,1,image,1,0)
plt.imshow(ref_line)
plt.title('Ref Line')
plt.show()


#Init Lists to hold the two clusters sorted according to the ref line
right_lines = []
left_lines = []

thresh = 0.7 #Dead zone threshold

for line in lines:
    
    #Check the position of the current line with respect to the ref line
    #If not append in the other cluster
    if line[0][0]< ref_pts[0]:
        
        #Check if the line is far enough from the ref    
        if line[0][0]/ref_pts[0] < thresh: 
            left_lines.append(line[0])
  
    elif line[0][0] > ref_pts[0]:
        
        if ref_pts[0]/line[0][0] > thresh: 
            right_lines.append(line[0])
        
        
#A list to hold both lanes lines

lanes = [left_lines,right_lines]  


#final filtered lines
filtered_lines = []

# There are two ways to fit a new line after separating each lane

##########################
######  METHOD 1 #########
##########################

#Forming new line from each cluster 
#by taking the min and max points of all cluster and fit a new line

for lane in lanes:
    
    #init a min and max points
    x_min = image.shape[1]
    x_max = 0

    #Loop over each cluster
    for line in lane:
        if line[0] < x_min:
            x_min = line[0]
            y_min = line[1]
        
        if line[2] > x_max:
            x_max = line[2]
            y_max = line[3]

    #Append the new fitted line             
    filtered_lines.append([x_min,y_min,x_max,y_max])

''' 
##########################
######  METHOD 2 #########
##########################
#Using the average of all points

for lane in lanes:
    
    #init x and y lists to separate x and y points for np.polyfit function
    x_min = 0
    y_min = 0
    x_max = 0
    y_max = 0

    
    #Loop over each cluster
    for line in lane:
        x_min += line[0]
        y_min += line[1]
        x_max += line[2]
        y_max += line[3]

    
    #Append the new fitted line       
    filtered_lines.append(list(map(int, [x_min/len(lane),y_min/len(lane),x_max/len(lane),y_max/len(lane)])))
''' 

#Draw The output lines
for line in filtered_lines:
    cv2.line(detected_lines,(line[0],line[1]),(line[2],line[3]),255,15)
    

plt.imshow(detected_lines)
plt.title('filtered lanes')
plt.show()
#fues the lines with the original image
final_im = cv2.addWeighted (detected_lines,1,image,1,0)

plt.imshow(final_im)
plt.title('Lane Detection')
plt.show()

#Save Image

#cv2.imwrite('test_images_output/solidWhiteRight.jpg', cv2.cvtColor(final_im, cv2.COLOR_RGB2BGR))  
````
